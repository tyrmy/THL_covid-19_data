# Turun siltojen liikennedata API

Demonstartion of open data analysis

```python
%matplotlib inline

import requests
from requests.exceptions import HTTPError, ConnectionError

import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# Fetch data from website

url = 'https://dev.turku.fi/datasets/ecocounter/2020/counters-15min.csv'
r = requests.get(url, timeout=5)
```

```python
# Print request status

print('Success %r' % r)
print('Elapsed time: ', r.elapsed)
print('Code status: ', r.reason)
```

```python
# Save response to a .csv file

with open('data.csv', 'wb') as file:
    file.write(r.content)
```

```python
# Generate a dataframe from csv file

data = pd.read_csv('./data.csv',  index_col='aika')
```

```python
# Plot values with time

fig, axes = plt.subplots(3,1, figsize=(14,9), sharex=True)

data.plot(y=['Auransilta AK','Auransilta AP'], ax=axes[0])
data.plot(y=['Auransilta JK','Auransilta JP'], ax=axes[1])
data.plot(y=['Auransilta PK','Auransilta PP'], ax=axes[2])

fig.autofmt_xdate()
```

```python
# Plot value frequency histogram

fig, axes = plt.subplots(3,1, figsize=(14,9))

data.plot.hist(y=['Auransilta AK','Auransilta AP'], ax=axes[0], bins=100)
data.plot.hist(y=['Auransilta JK','Auransilta JP'], ax=axes[1], bins=100)
data.plot.hist(y=['Auransilta PK','Auransilta PP'], ax=axes[2], bins=100)
```

# THL Open  Covid data

Demonstration of open data provided by Terveydenhuoltolaitos.

First we plot the age distribution of infected patients.

https://sampo.thl.fi/pivot/prod/fi/epirapo/covid19case/fact_epirapo_covid19case.json?column=ttr10yage-444309

```python
%matplotlib inline

import requests
from requests.exceptions import HTTPError, ConnectionError

import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

import json
```

```python
url = 'https://sampo.thl.fi/pivot/prod/fi/epirapo/covid19case/fact_epirapo_covid19case.json?column=ttr10yage-444309'

# Set custom headers
headers = {'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.36'}
r = requests.get(url, timeout=5, headers=headers)

# Print request status
print('Success %r' % r)
print('Elapsed time: ', r.elapsed)
print('Code status: ', r.reason)
```

```python
# Save response to a .csv file

with open('covid-age.json', 'wb') as file:
    file.write(r.content)
```

```python
# Load file

with open('covid-age.json', 'r') as file:
    json_data = file.read()
```

```python
# Get wanted values from json file

loaded_json = json.loads(json_data)

infected = list(loaded_json['dataset']['value'].values())
age = list(loaded_json['dataset']['dimension']['ttr10yage']['category']['label'].values())

print(age)
print(infected)
```

```python
df = pd.DataFrame(list(zip(age,infected)), columns=['age', 'infected'])
df.drop(df.tail(1).index,inplace=True)
print(df)
```

```python
df = df.astype({'infected': 'int32'})
print(df.dtypes)
```

```python
df.plot.bar(x='age', y='infected', figsize=(10,6), rot=45, title='Age distribution of confirmed infection cases')
```

## Ditribution by municipal regions

Second plot is the regional infection distribution. All steps of code are gathered in one cell.

https://sampo.thl.fi/pivot/prod/fi/epirapo/covid19case/fact_epirapo_covid19case.json?column=hcdmunicipality2020-445222

```python
url = 'https://sampo.thl.fi/pivot/prod/fi/epirapo/covid19case/fact_epirapo_covid19case.json?column=hcdmunicipality2020-445222'

# Set custom headers
headers = {'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.36'}
r = requests.get(url, timeout=5, headers=headers)

# Print request status
print('Success %r' % r)
print('Elapsed time: ', r.elapsed)
print('Code status: ', r.reason)

# Save response to a .csv file
with open('covid-region.json', 'wb') as file:
    file.write(r.content)
    
# Load file
with open('covid-region.json', 'r') as file:
    json_data = file.read()
    
# Get wanted values from json file
loaded_json = json.loads(json_data)

infected = list(loaded_json['dataset']['value'].values())
municipal = list(loaded_json['dataset']['dimension']['hcdmunicipality2020']['category']['label'].values())

# Make dataframe and convert strings to int
df = pd.DataFrame(list(zip(municipal,infected)), columns=['municipal', 'infected'])
df.drop(df.tail(1).index,inplace=True)
df = df.astype({'infected': 'int32'})

# Sort values by the amount of infectd patients
df.sort_values(by='infected', inplace=True)

df.plot.bar(x='municipal', y='infected', figsize=(14,8), rot=70, title='Municipal region infection confirms')
```

## Daily infections

Third plot is the daily infection confirms.

https://sampo.thl.fi/pivot/prod/fi/epirapo/covid19case/fact_epirapo_covid19case.json?column=dateweek2020010120201231-443702L

```python
url = 'https://sampo.thl.fi/pivot/prod/fi/epirapo/covid19case/fact_epirapo_covid19case.json?column=dateweek2020010120201231-443702L'

# Set custom headers
headers = {'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.36'}
r = requests.get(url, timeout=5, headers=headers)

# Print request status
print('Success %r' % r)
print('Elapsed time: ', r.elapsed)
print('Code status: ', r.reason)

# Save response to a .csv file
with open('covid-daily.json', 'wb') as file:
    file.write(r.content)
    
# Load file
with open('covid-daily.json', 'r') as file:
    json_data = file.read()
    
# Get wanted values from json file
loaded_json = json.loads(json_data)

infected = list(loaded_json['dataset']['value'].values())
daily = list(loaded_json['dataset']['dimension']['dateweek2020010120201231']['category']['label'].values())

# Make dataframe and convert strings to int
df = pd.DataFrame(list(zip(daily,infected)), columns=['datetime', 'infected'])
df = df.astype({'infected': 'int32'})
df['datetime'] = pd.to_datetime(df['datetime'], format="%Y-%m-%d")
df['cumsum'] = df['infected'].cumsum()

# Sort values by the amount of infected patients
df.sort_values(by='infected', inplace=True)

df.plot(x='datetime', y='infected', figsize=(12,5), rot=70, title='Daily infection confirms')
df.plot(x='datetime', y='cumsum', figsize=(12,5), rot=70, title='Sum of confirms')
print('Total infections: ', df['cumsum'].max())
```
